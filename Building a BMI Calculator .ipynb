## Building a BMI Calculator
This project computes users’ BMI values and assigns them to corresponding health categories.

The Body Mass Index (BMI) is calculated using the following formula:
$$
\mathrm{BMI} = \frac{\text{Weight (lb)} \times 703}{(\text{Height (in)})^2}
$$

#### BMI Categories (World Health Organization Standard):
| Category | BMI Range |
|-----------|------------|
| Underweight | < 18.5 |
| Normal weight | 18.5 – 24.9 |
| Overweight | 25.0 – 29.9 |
| Obesity (Class I) | 30.0 – 34.9 |
| Obesity (Class II) | 35.0 – 39.9 |
| Obesity (Class III) | ≥ 40.0 |

```python
# Greet the User.

Name = str(input("Enter your name:"))
print(f"Hello {Name}, Welcome to your BMI calculator!")
print("I would need the following details from you.")
```

```python
# User input

Weight = float(input('Enter your weight in pounds here:'))
Height = float(input('Enter your height in inches here:'))

#BMI Calculator
BMI = (Weight * 703)/(Height * Height)

# BMI Categorization
if BMI > 0:
    if (BMI < 18.5):
        category = "Underweight"
        advice = ("You’re underweight — maybe add an extra snack or two.\n"
                  "On a serious note, try including more nutritious and calorie-rich foods in your diet.")
    elif (BMI <= 24.9):
        category = "Normal weight"
        advice = ("You’re right where you need to be — keep doing what you’re doing!\n"
                  "Maintain your healthy weight through balanced nutrition and regular physical activity.")
    elif (BMI <= 29.9):
        category = "Overweight"
        advice = ("You’re a bit on the heavier side — maybe skip that extra midnight snack.\n"
                  "A little more movement and mindful eating can go a long way.")
    elif (BMI <= 34.9):
        category = "Obesity (Class I)"
        advice =  ("You’re in the obese range — your body deserves some extra care.\n"
                   "It might be beneficial to consult a healthcare professional for personalized guidance on weight management.")
    elif (BMI <= 39.9):
        category = "Obesity (Class II)"
        advice = ("You’re in the obese range — your body deserves some extra care.\n"
                  "It might be beneficial to consult a healthcare professional for personalized guidance on weight management.")
    elif (BMI >= 40):
        category = "Obesity (Class III)"
        advice = ("You’re in the obese range — your body deserves some extra care.\n"
                  "It might be beneficial to consult a healthcare professional for personalized guidance on weight management.")
    else:
        print(f"{Name}, kindly enter valid inputs.")
```

```python
# results
print(f"\nHi {Name}, your BMI is {BMI:.2f}.")
print(f"Based on the BMI scale, you are categorized as: {category}.")
print(f"Health Tip:\n{advice}")
print("Thank you for using the BMI calculator!")
```
